```python
import json

from matplotlib import pyplot as plt
# matplotlib configuration
plt.rcParams["figure.figsize"] = (8, 8)
plt.rcParams["axes.labelsize"] = 14
plt.rcParams["axes.labelweight"] = "bold"
```

```python
largest_size = 799_997
half = largest_size // 2
fourth = half // 2
eighth = fourth // 2
all_sizes = [eighth, fourth, half, largest_size]
times = [2150, 4544, 8987, 19228]
```

```python
x = all_sizes
y = times
plt.figure()
plt.xlabel("Number of documents")
plt.ylabel("Time (seconds)")
plt.plot(x, y)
plt.scatter(x, y, c="r")
plt.savefig("img/perf_indexer.png")
```

### Distribution of sizes of inverted lists

```python
# We assume the statistics file has been created already
stats_fpath = "statistics.json"
posting_lens = json.loads(open(stats_fpath).read())["Distribution of posting lengths"]
```

```python
plt.figure()
plt.xlabel("Size of inverted list")
plt.ylabel("Number of inverted lists")
plt.hist(posting_lens, bins=25, log=True, edgecolor="w")
plt.savefig("img/distribution_inverted_lists.png")
```
